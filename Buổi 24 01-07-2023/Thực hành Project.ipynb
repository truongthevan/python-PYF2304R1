```python
class Bookstore:
    def __init__(self, name, category, price):
        self.name = name
        self.category = category
        self.price = price
    
    def __str__(self):
        return f"name: {self.name}, category: {self.category}, price: {self.price}"
```

```python
from tkinter import *
from tkinter import messagebox
from tkinter.ttk import *

window = Tk()
window.title("Management the Bookstore")

book_library = []

def add_book(): # chức năng thêm sách
    name_value = name.get()
    category_value= category.get()
    price_value = price.get()
    if name_value == "" or category_value =="" or price_value =="":
        messagebox.showerror("book", "please enter name, category and price")
    else:
        new_book = Bookstore(name_value, category_value,int(price_value))
        book_library.append(new_book)

        print([book.name for book in book_library])
        messagebox.showinfo("book", f"Added {name_value} successfully")
        
def search_book():# chức năng tìm sách
    search_data = search1.get()
    check = False
    for i in range(len(book_library)):
        if book_library[i].name == search_data:
            messagebox.showinfo("Search book", book_library[i])           
            check = True
            break
    if check == False:
        messagebox.showerror("Search book", "can't find book")

def delete_book():#chức năng xóa sách
    delete_name = search1.get()
    check = False
    for i in range(len(book_library)):
        if book_library[i].name == delete_name:
            del book_library[i]
            print([book.name for book in book_library])
            messagebox.showinfo("Delete book", f"Delete {delete_name} sucessful")           
            check = True
            break
    if check == False:
        messagebox.showerror("Delete book", "can't find book")
        
def update_book():#chức năng cập nhật giá tiền
    update_name = search1.get()
    check = False
    for i in range(len(book_library)):
        if book_library[i].name == update_name:
            book_library[i].price = price.get()
            print([book.name for book in book_library])
            messagebox.showinfo("Update book", f"update {update_name} sucessful")           
            check = True
            break
    if check == False:
        messagebox.showerror("update book", "can't find book")        
        
# tạo label
lbl_title = Label(window, text=" The Bookstore Management System", anchor="w", justify="left")
lbl_title.grid(row=0, column=0, columnspan=3)
lbl_name = Label(window, text="Book name")
lbl_name.grid(row=1, column=0)
lbl_category = Label(window, text="Book category")
lbl_category.grid(row=2, column=0)
lbl_price = Label(window, text="Book's price")
lbl_price.grid(row=3, column=0)
lbl_search = Label(window, text="Search")
lbl_search.grid(row=1, column=3)

#tạo Entry
name = Entry(window)
name.grid(row=1, column=1)
price = Entry(window)
price.grid(row=3, column=1)
search1 = Entry(window)
search1.grid(row=1, column=4)


# tạo button
add = Button(window, text="Add", command=add_book)
add.grid(row = 5, column =0)
search = Button(window, text="Search", command=search_book)
search.grid(row=2, column=3)
delete = Button(window,text="Delete", command=delete_book)
delete.grid(row=2, column=4)
update = Button(window, text="Update price", command=update_book)
update.grid(row=7, column=0)


# tạo combobox
category = Combobox(window)
#Các giá trị của hộp chọn
category['values']= ("Science", "Economy", "Engineering", "Technology", "Law", "Mom and Baby", "Dictionary", "English")
category.grid(row=2,column=1)


window.mainloop()
```
