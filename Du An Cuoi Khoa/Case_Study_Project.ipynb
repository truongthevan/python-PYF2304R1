```python
class Footballteam:
    '''Lớp đội bóng, quản lý cầu thủ theo tên, tuổi, chiều cao, cân nặng, và vị trí thi đấu trên sân'''

    def __init__(self, name, age, height, weight, position):
        '''hàm khởi tạo mặc định tham số'''
        self.name = name
        self.age = age
        self.height = height
        self.weight = weight
        self.position = position

    def __str__(self):
        '''hàm khởi tạo mặc định in thông tin'''
        return f"Tên: {self.name}, Tuổi: {self.age}, Chiều cao: {self.height}, Cân nặng: {self.weight}, Vị trí: {self.position}"


from tkinter import *
from tkinter import messagebox
from tkinter.ttk import *
import tkinter as tk
from tkinter import ttk
from tkinter.messagebox import showinfo

window = Tk()
window.title("Trình Quản Lý Cầu Thủ")
window.configure(bg="#6bcce0")
window.geometry("1010x250") 

player_list = []

def add_player():  # chức năng thêm cầu thủ, cần cảnh báo khi tên đã tồn tại
    '''hàm thực hiện chức năng thêm thông tin cầu thủ'''
    check = False
    for i in range(len(player_list)):
        name_value = name.get()
        if name_value == player_list[i].name:
            messagebox.showwarning("Cầu thủ", f"Đã tồn tại {name_value}")
            check = True
            break
    if check == False:
        name_value = name.get()
        age_value = age.get()
        height_value = height.get()
        weight_value = weight.get()
        position_value = position.get()

        if name_value == "" or age_value == "" or height_value == "" or weight_value == "" or position_value == "":
            messagebox.showerror("cầu thủ", "Vui lòng nhập tên, tuổi, chiều cao, cân nặng, vị trí thi đấu")
        #elif name_value in player_list.name == True:
            #messageshowwarning("Cầu thủ", f"Đã tồn tại {name_value}")
        else:
            new_player = Footballteam(name_value, float(age_value), float(height_value), float(weight_value),
                                      position_value)
            player_list.append(new_player)
            print([player.name for player in player_list])
            messagebox.showinfo("Cầu thủ", f"Đã thêm cầu thủ {name_value} thành công")
            for i in range(len(player_list)):#append danh sách vào treeview
                tree_list = []
                tree_list.append((player_list[i].name, player_list[i].age, player_list[i].height, player_list[i].weight,
                                  player_list[i].position))
            for player in tree_list:
                tree.insert("", "end", values=player)

def search_player():  # chức năng thêm cầu thủ
    '''hàm thực hiện chức năng tìm kiếm cầu thủ trong thư viện'''
    if name.get() == "":
        messagebox.showerror("Cầu thủ", "Nhập tên vào ô Tên Cầu Thủ")
    else:
        search_data = name.get()
        check = False
        for i in range(len(player_list)):
            if search_data == player_list[i].name:
                messagebox.showinfo("Cầu thủ", player_list[i])
                check = True
                break
        if check == False:
            messagebox.showerror("cầu thủ", f"Cầu thủ {search_data} không tồn tại")

def change_player(): # chức năng đổi thông tin cầu thủ
    '''hàm thực hiện chức năng đổi thông tin cầu thủ'''
    if name.get() == "":
        messagebox.showerror("Cầu thủ", "Hãy tìm kiếm để đảm bảo cầu thủ có trong dánh sách")
    else:
        update_player = name.get()
        check = False
        for i in range(len(player_list)):
            if player_list[i].name == update_player:
                if age.get() == "":
                    pass
                else:
                    player_list[i].age = float(age.get())
                if height.get() == "":
                    pass
                else:
                    player_list[i].height = float(height.get())
                if weight.get() == "":
                    pass
                else:
                    player_list[i].weight = float(weight.get())
                if position.get() == "":
                    pass
                else:
                    player_list[i].position = position.get()
                check = True
                messagebox.showinfo("cầu thủ", f"Cầu thủ {update_player} đổi thông tin thành công")
                break
        if check == False:
            messagebox.showerror("cầu thủ", f"Cầu thủ {update_player} không tồn tại")
    tree_list = []
    for i in range(len(player_list)): #update danh sách trong treeview
        tree_list.append((player_list[i].name, player_list[i].age, player_list[i].height, player_list[i].weight,
                          player_list[i].position))
    for item in tree.get_children():
        tree.delete(item)
    for player in tree_list:
        tree.insert("", "end", values=player)


def delete_player():#chức năng xóa cầu thủ trong danh sách
    '''hàm thực hiện chức năng xóa thông tin cầu thủ'''
    if name.get() == "":
        messagebox.showerror("Cầu thủ", "Hãy tìm kiếm để đảm bảo cầu thủ có trong dánh sách")
    else:
        delete_player = name.get()
        check = False
        for i in range(len(player_list)):
            if delete_player == player_list[i].name:
                del player_list[i]
                print([player.name for player in player_list])
                messagebox.showinfo("cầu thủ", f"xóa {delete_player} thành công")
                check = True
                break
        if check == False:
            messagebox.showerror("cầu thủ", f"Cầu thủ {delete_player} không tồn tại")
    tree_list = []
    for i in range(len(player_list)): # xóa khỏi treeview
        tree_list.append((player_list[i].name, player_list[i].age, player_list[i].height, player_list[i].weight,
                          player_list[i].position))
    for item in tree.get_children():
        tree.delete(item)
    for player in tree_list:
        tree.insert("", "end", values=player)

# tạo file và lưu danh sách
import csv

def save_list():#chức năng lưu hết danh sách
    '''Hàm thực hiện chức năng lưu danh sách cầu thủ'''
    file = open("players.csv", mode="w")
    file.write("")
    file.close()
    players = []
    for i in range(len(player_list)):
        player = Footballteam(player_list[i].name, player_list[i].age, player_list[i].height, player_list[i].weight,
                              player_list[i].position)
        players.append(player)
    for player in players:
        data_dict = player.__dict__
        #print(data_dict)
        header = list(data_dict.keys())
        with open("players.csv", mode="a", newline="") as csvfile:  # 'a' for append; 'w' for write; 'r' for read
            writer = csv.DictWriter(csvfile,  fieldnames=header, dialect="excel")
            # print(csvfile.tell())
            if csvfile.tell() == 0:  # check if the file is empty
                writer.writeheader()
            writer.writerow(data_dict)
    messagebox.showinfo("cầu thủ", "đã lưu danh sách cầu thủ")

def load_list():#chức năng tải hết danh sách đã lưu
    '''Hàm thực hiện chức năng tải lên danh sách cầu thủ'''
    player_list.clear()# vị trí này đang conflict
    for item in tree.get_children():
        tree.delete(item)        
    file = open("players.csv")
    csvreader = csv.reader(file)
    next(csvreader) #bỏ dòng đầu tiên trong file CSV, do đã có header
    for row in csvreader:
        tree.insert("", "end", values=row)
        data = row #.split(',') chưa hiểu vì sao cần bỏ slpit
        name = data[0]
        age = data[1]
        height = data[2]
        weight = data[3]
        position = data[4]
        item = Footballteam(name, age, height, weight, position)
        player_list.append(item) #tao object them vao list player_list
    file.close()
    messagebox.showinfo("cầu thủ", "đã tải lên danh sách")

def clear_list():#chức năng xóa hết dánh sách
    '''Hàm thực hiện chức xóa hết danh sách cầu thủ'''
    for item in tree.get_children():
        tree.delete(item)
    player_list.clear()
    file = open("players.csv", mode="w")
    file.write("")
    file.close()
    messagebox.showwarning("Cầu thủ", "Danh sách cầu thủ đã xóa hết!")

# xóa các dư liệu đang hiện hành trên màn hình nhập vào
def reset_infor():
    '''Hàm reset Entry input'''
    name.delete(0, END)
    age.delete(0, END)
    height.delete(0, END)
    weight.delete(0, END)
    position.delete(0, END)

# tạo label
lbl_title = Label(window, text="Quản Lý Thông Tin", font=("time, 12"))
lbl_title.grid(row=0, column=0, columnspan=4)
lbl_name = Label(window, text="Tên Cầu Thủ")
lbl_name.grid(row=1, column=0)
lbl_age = Label(window, text="Tuổi Cầu Thủ")
lbl_age.grid(row=2, column=0)
lbl_height = Label(window, text="Chiều Cao (M)")
lbl_height.grid(row=3, column=0)
lbl_weight = Label(window, text="Cân Nặng (Kg)")
lbl_weight.grid(row=4, column=0)
lbl_position = Label(window, text="Vị Trí", anchor="w", justify="left")
lbl_position.grid(row=5, column=0)
lbl_info = Label(window, text="Danh Sách Cầu Thủ", font=("time, 10"))
lbl_info.grid(row=0, column=4)

# tạo Entry
name = Entry(window, width=22)
name.grid(row=1, column=1)
age = Entry(window, width=22)
age.grid(row=2, column=1)
height = Entry(window, width=22)
height.grid(row=3, column=1)
weight = Entry(window, width=22)
weight.grid(row=4, column=1)

# tạo combobox vì vị trí chiến thuật trên sân là cố định trong bóng đá
position = Combobox(window)
position['value'] = ("", "Tien Dao Cam", "Tien Dao Canh Trai", "Tien Dao Canh Trai", "Tien Ve Cong", "Tien Ve Trung Tam",
                    "Tien Ve Trai", "Tien Ve Phai", "Trung Ve", "Hau Ve Bien Trai", "Hau Ve Bien Phai", "Thu Mon")
position.grid(row=5, column=1)

# tạo button
add = Button(window, text="Thêm", command=add_player)
add.grid(row=1, column=2)
search = Button(window, text="Tìm Kiếm", command=search_player)
search.grid(row=2, column=2)
update = Button(window, text="Thay Đổi", command=change_player)
update.grid(row=3, column=2)
delete = Button(window, text="Xóa", command=delete_player)
delete.grid(row=4, column=2)
reset = Button(window, text="Nhập Lại", command=reset_infor)
reset.grid(row=5, column=2)
save = Button(window, text="Lưu Danh Sách", width=15, command=save_list)
save.grid(row=1, column=5)
get = Button(window, text="Tải Danh Sách", width=15, command=load_list)
get.grid(row=2, column=5)
deleteall = Button(window, text="Xóa Danh Sách", width=15, command=clear_list)
deleteall.grid(row=3, column=5)

# tạo treeview-định nghĩa cột
columns = ("name", "age", "height", "weight", "position")
tree = ttk.Treeview(window, columns=columns, show='headings')
# định nghĩa tiêu đề
tree.heading("name", text="Tên")
tree.heading("age", text="Tuổi")
tree.heading("height", text="Chiều cao (M)")
tree.heading("weight", text="Cân nặng (Kg)")
tree.heading("position", text="Vị Trí")
tree.column("name", width=200)
tree.column("age", width=50)
tree.column("height", width=100)
tree.column("weight", width=100)
tree.column("position", width=150)
tree.grid(row=1, rowspan=6, column=4, sticky='nsew')

# thêm scrollbar
scrollbar = ttk.Scrollbar(window, orient=tk.VERTICAL, command=tree.yview)
tree.configure(yscroll=scrollbar.set)
scrollbar.place(x=885, y=195)

window.mainloop()
```
