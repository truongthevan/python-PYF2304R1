### Bài 10:
Tại một cửa hàng nọ có menu như sau:

1. Gà rán giá 35k
2. Hamburger giá 50k
3. Pesi giá 10k

Nhập số lượng món hàng cần mua

Tính và in ra tổng giá tiền

Nếu giá trị đơn hàng trên 2 triệu, giảm 10% cho khách, trên 4 triệu, giảm 15%

Tính và in ra tổng giá tiền được giảm

Tính và in ra tổng giá tiền khách cần trả

In ra dòng chữ "Cảm ơn quý khách đã mua hàng"

````python
"""Ví dụ:
Nhập số lượng gà rán khách mua:100
Nhập số lượng hamburger khách mua:100
Nhập số lượng pesi khách mua:15
Tổng giá tiền đơn hàng là: 8.650.000
Tổng tiền được giảm là: 1.297.500
Tổng tiền phải thanh toán: 7.352.500
Cảm ơn quý khách đã mua
"""
````

````python
# so luong ga ran mua
ga_ran = int(input("so phan ga ran da mua: "))
# so luong ga ran mua
hamburger = int(input("so phan hamburger da mua: "))
# so luong pepsi mua
pepsi = int(input("so phan pepsi da mua: "))

# ga ran 35k, hamburger 50k, pepsi 10k
total = (ga_ran*35000) + (hamburger*50000) + (pepsi*10000)
giam_gia_1 = total*10/100 # giam gia 10%
phai_tra_1 = total - giam_gia_1
giam_gia_2 = total*15/100 # giam gia 15%
phai_tra_2 = total - giam_gia_2
print("tong so tien da mua: ", end="") # thu cach in noi nhau
print(total)

if total < 2000000:
    print("cam con quy khach da mua hang")
elif total > 2000000 and total < 4000000:
    print(f"so tien duoc giam: {giam_gia_1}")
    print(f"so tien phai thanh toan: {phai_tra_1}")
    print("cam con quy khach da mua hang")
else:
    print(f"so tien duoc giam: {giam_gia_2}")
    print(f"so tien phai thanh toan: {phai_tra_2}")
    print("cam con quy khach da mua hang")
````
